# Train Test Split

```python
df.set_index("Date",inplace=True)
df
```

```python
x=df[['Open','High','Low','Volume']]
y=df['Close']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=False,random_state=42)
```

# Normalization the values

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# Model creation

__linear regression model__

```python
from sklearn.linear_model import LinearRegression # Use Ridge regression for regularization
from sklearn.metrics import mean_absolute_error, mean_squared_error


model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)

# Make predictions using the best model
pred = model.predict(x_test)

mse=np.round(mean_squared_error(y_test,pred),2)
mae=np.round(mean_absolute_error(y_test,pred),2)
print (mse)
print (mae)
```

```python
fig=plt.figure(figsize=(15,8))
p=pd.Series(pred,index=y_test.index)
plt.plot(y_test)
plt.plot(p)
plt.legend(['actual','predict'])
plt.title("pred vs actual values")
plt.show()
```

__SVR__

```python
from sklearn.svm import SVR
model=SVR(C=4.0,epsilon=0.00002)
model.fit(x_train,y_train)
pred=model.predict(x_test)

mse=np.round(mean_squared_error(y_test,pred),2)
mae=np.round(mean_absolute_error(y_test,pred),2)
print(mse)
print (mae)
```

```python
fig=plt.figure(figsize=(15,8))
p=pd.Series(pred,index=y_test.index)
plt.plot(y_test)
plt.plot(p)
plt.legend(['actual','predict'])
plt.title("pred vs actual values")
plt.show()
```

# LSTM 

```python
x=df[['Close']]
length=100

# split dataset (training,validation,testing)
train_set=x.iloc[:800].values
validation_set=x.iloc[800:1000].values
test_set=x.iloc[1000:].values

# features scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
train_set_scaled = scaler.fit_transform(train_set)
validation_set_scaled = scaler.transform(validation_set)
test_set_scaled = scaler.transform(test_set)


print (train_set_scaled.shape)
print (validation_set_scaled.shape)
print (test_set_scaled.shape)
```

__Creating a data structure with 50 time step and 1 output__

```python
# for training set

x_train=[]
y_train=[]

for i in range(length,len(train_set)):
    x_train.append(train_set_scaled[i-length:i,0])
    y_train.append(train_set_scaled[i,0])
    print (i-length,"i-1",i-1,"i",i)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

```python
# for validation set
x_val=[]
y_val=[]

for i in range(length,len(validation_set)):
    x_val.append(validation_set_scaled[i-length:i,0])
    y_val.append(validation_set_scaled[i,0])


x_val=np.array(x_val)
y_val=np.array(y_val)
x_val=np.reshape(x_val,(x_val.shape[0],x_val.shape[1],1))
```

```python
# for test set
x_test=[]
y_test=[]

for i in range(length,len(test_set)):
    x_test.append(test_set_scaled[i-length:i,0])
    y_test.append(test_set_scaled[i,0])


x_test=np.array(x_test)
y_test=np.array(y_test)
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

__Build model__

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential()
# Adding first LSTM layer and Dropout
model.add(LSTM(units=128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))  # Increase dropout rate

# Adding second LSTM layer and Dropout
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(0.2))  # Increase dropout rate

# Adding third LSTM layer and Dropout
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(0.2))  # Increase dropout rate

# # Adding fourth LSTM layer and Dropout
# model.add(LSTM(units=128, return_sequences=True))
# model.add(Dropout(0.2))  # Increase dropout rate

# # Adding fifth LSTM layer and Dropout
# model.add(LSTM(units=128, return_sequences=True))
# model.add(Dropout(0.2))  # Increase dropout rate

# Adding sixth LSTM layer and Dropout
model.add(LSTM(units=128, return_sequences=False))
model.add(Dropout(0.2))  # Increase dropout rate


# Adding the output layer
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)

# Train the model with early stopping
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test), callbacks=[early_stopping])
```

```python
loss=pd.DataFrame(model.history.history)
fig=plt.figure(figsize=(15,8))
plt.plot(loss)
plt.title("val loss VS triaining loss")
plt.legend(['val','training'])
plt.show()
```

```python
# prediction values
pred=model.predict(x_test)
test=pd.DataFrame(columns=['test','pred'])
test['test']=y_test
test['pred']=pred.flatten()
test
```

```python
from sklearn.metrics import r2_score

fig=plt.figure(figsize=(15,8))
plt.title("Test VS Predict value",size=20,weight='bold')
plt.plot(y_test)
plt.plot(pred)
plt.legend(['test','pred'])

r2=np.round(r2_score(y_test,pred),2)
mse=np.round(mean_squared_error(y_test,pred),2)
mae=np.round(mean_absolute_error(y_test,pred),2)

print (f"r2 = {r2}")
print (f"mse = {mse}")
print (f"mae = {mae}")
```
